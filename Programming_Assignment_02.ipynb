```python
a = int(input("Enter Kilometers :  "))
b = (a / 1.609)
print(f"{a} kilometers mentioned are {b} miles")
```

```python
# 2.Write a Python program to convert Celsius to Fahrenheit?
```

```python
c = int(input("Enter Celsius : "))
d = (c * (9/5)) + 32
print(c , "Degree Celsius is " , d , "Degree Fahrenheit")
```
